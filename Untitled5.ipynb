```python
class student:
  def __init__(self , name , age , marks):
    self.name = name
    self.age = age
    self.marks = marks


  def display_info(self):
      print(f"Name: {self.name} , age: {self.age} , marks{self.marks}")
s1 = student("ali" , 21 , 56)
s1.display_info()
```

```python
class teacher:

  def __init__(this , name , subject ,address):
    this.name = name
    this.subject = subject
    this.address = address

  def display_info(this):
    print(f"teacher name: {this.name}")
    print(f"teacher subject {this.subject}")
    print(f"teacher address {this.address}")

t1 = teacher("ayesha" , "english" , "gujranwala")
t1.display_info()



```
